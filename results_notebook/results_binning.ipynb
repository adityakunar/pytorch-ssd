## Plotting results abut different object sizes

```python
import matplotlib.pyplot as plt
```

```python
bins = [
    "0-5%",
    "5-10%",
    "10-20%",
    "20-40%",
    "40-60%",
    "60-80%",
    "80-100%"
]
```

```python
mbnet_augm = [
    0.331,
    0.762,
    0.830,
    0.821,
    0.869,
    0.889,
    0.888
]

vgg16_augm = [0.571,0.881,0.861,0.873,0.918,0.917,0.893]

mbnet_no_augm = [0.216,0.597,0.746,0.706,0.792,0.822,0.787]

vgg16_no_augm = [0.315,0.632,0.750,0.714,0.744,0.786,0.783]
```

```python
plt.plot(bins, vgg16_augm, label="VGG16 w/ augmentation")
plt.plot(bins, mbnet_augm, label="MobileNet w/ augmentation")
plt.plot(bins, mbnet_no_augm, label="MobileNet w/out augmentation")
plt.plot(bins, vgg16_no_augm, label="VGG16 w/out augmentation")

plt.xlabel("Bins")
plt.ylabel("Prediction accuracy")

plt.legend()
plt.show()
```
